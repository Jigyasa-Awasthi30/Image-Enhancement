# FLIPPING AN IMAGE

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("y direction")
flip=cv2.flip(image,0)
cv2_imshow(flip)
# 0 represent y direction
# 1 represent x direction
# 1 represent x and y (both)direction
print("x direction")
flip1=cv2.flip(image,1)
cv2_imshow(flip1)
print("x and y (both)direction")
flip2=cv2.flip(image,-1)
cv2_imshow(flip2)
```

# BITWISE OPERATION

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
print("RECTANGLE")
rectangle=np.zeros((300,300),dtype="uint8")
cv2.rectangle(rectangle,(25,25),(275,275),255,-1)
cv2_imshow(rectangle)
print("CIRCLE")
circle=np.zeros((300,300),dtype="uint8")
cv2.circle(circle,(150,150),150,255,-1)
cv2_imshow(circle)
print("OR Operation")
bitwiseor=cv2.bitwise_or(rectangle,circle)
cv2_imshow(bitwiseor)
print("AND Operation")
bitwiseand=cv2.bitwise_and(rectangle,circle)
cv2_imshow(bitwiseand)
print("XOR Operation")
bitwisexor=cv2.bitwise_xor(rectangle,circle)
cv2_imshow(bitwisexor)
print("NOT Operation")
bitwisenot=cv2.bitwise_not(circle)
cv2_imshow(bitwisenot)
```
