# SMOOTHING ANG BLURRING

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("Average Blurring")
blurred=cv2.blur(image,(5,5))
cv2_imshow(blurred)
print("Gaussian Blurring")
gaussian=cv2.GaussianBlur(image,(5,5),0)
cv2_imshow(gaussian)
print("Median Blurring")
median=cv2.medianBlur(image,3)
cv2_imshow(median)
```

# MASKING

**Image masking is a photo editing technique that separates specific areas of an image from the rest, allowing for more precise edits. It's a non-destructive method, meaning the original image remains intact while adjustments are made.**

***RECTANGULAR*** ***MASKING***

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("mask")
mask=np.zeros(image.shape[:2],dtype="uint8")
cv2.rectangle(mask,(100,90),(300,300),255,-1)
cv2_imshow(mask)
print("masked image")
masked=cv2.bitwise_and(image,image,mask=mask)
cv2_imshow(masked)
```

***CIRCULAR*** ***MASKING***

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
print("mask")
mask=np.zeros(image.shape[:2],dtype="uint8")
cv2.circle(mask,(150,150),150,255,-1)
cv2_imshow(mask)
print("masked image")
masked=cv2.bitwise_and(image,image,mask=mask)
cv2_imshow(masked)
```
