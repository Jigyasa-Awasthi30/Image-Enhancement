# THRESHOLDING THE IMAGE

**Image thresholding is a technique that converts a grayscale or color image into a binary image, separating foreground and background pixels**

```python
import cv2
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/natureimage.jpg")
blurred=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
print("threshold image")
(T,thresh)=cv2.threshold(blurred,200,255,cv2.THRESH_BINARY)
cv2_imshow(thresh)
print("threshold image inverted")
(T,threshinv)=cv2.threshold(blurred,200,255,cv2.THRESH_BINARY_INV)
cv2_imshow(threshinv)
print(" Masking the image and the threshold image inverted")
cv2_imshow(cv2.bitwise_and(image,image,mask=threshinv))

```

**otsu** **thresholding**

```python
(T,threshinv)=cv2.threshold(blurred,155,255,cv2.THRESH_BINARY_INV|cv2.THRESH_OTSU)
cv2_imshow(threshinv)
```

# CONTOUR


**Image contouring is a technique that identifies and outlines the boundaries of objects in an image**

```python
import cv2
import imutils
from google.colab.patches import cv2_imshow
image=cv2.imread("/content/drive/MyDrive/Colab Notebooks/depositphotos_188711700-stock-photo-abstract-illustration-nine-different-shapes.jpg")
cv2_imshow(image)
```

**Internal contouring**

```python
import cv2
import imutils
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/depositphotos_188711700-stock-photo-abstract-illustration-nine-different-shapes.jpg")
cv2_imshow(image)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

cnts = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
clone = image.copy()
cv2.drawContours(clone, cnts, -1, (0, 0, 255), 2)
cv2_imshow(clone)
```

**External Contouring**

```python
import cv2
import imutils
from google.colab.patches import cv2_imshow
image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/depositphotos_188711700-stock-photo-abstract-illustration-nine-different-shapes.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
clone = image.copy()
cv2.drawContours(clone, cnts, -1, (0, 0, 255), 2)
cv2_imshow(clone)
```
