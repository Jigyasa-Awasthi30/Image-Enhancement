# Drawing Function

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
canvas=np.ones((500,500,3),dtype="uint8")*255
cv2_imshow(canvas)
black=(0,0,0)
cv2.line(canvas,(0,0),(400,400),black,3)
cv2_imshow(canvas)
red=(0,0,255)
cv2.rectangle(canvas,(150,100),(100,255),red,5)
cv2_imshow(canvas)
green=(0,255,0)
cv2.circle(canvas,(100,100),50,green,-1)
cv2_imshow(canvas)
```

# FOR Random Circle

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
canvas=np.zeros((500,500,3),dtype="uint8")
for i in range(0,25):
  radius=np.random.randint(5,200)
  color=np.random.randint(0,255,size=(3,)).tolist()
  pt=np.random.randint(0,400,size=(2,))
  cv2.circle(canvas,tuple(pt),radius,color,-1)
cv2_imshow(canvas)
```
